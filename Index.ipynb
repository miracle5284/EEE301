# Welcome to Jupyter!

```python
from matplotlib import pyplot as plt
%matplotlib Notebook
import numpy as np
vd = np.arange(-306, 0, 1)
f = [1/(2*np.pi* np.sqrt(25.4*10**(-20)/np.sqrt((1-(v_d/.6))))) for v_d in vd]
plt.plot(vd, f)
plt.title('Graph of Vd against Frequency')
plt.ylabel('Frequency')
plt.xlabel('Vd')
plt.show()
```
